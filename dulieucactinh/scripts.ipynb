```python
import random

def extend_years(
    df,
    start_year=2024,
    end_year=2030
):

    for year in range(
        start_year,
        end_year + 1
    ):

        prev_year = year - 1

        if prev_year not in df.columns:

            raise Exception(
                f"Missing column {prev_year}"
            )

        df[year] = df[prev_year].apply(

            lambda x:

            round(

                x *

                (

                    1 +

                    random.triangular(

                        -0.03,

                        0.05,

                        0.03

                    )

                ),

                1

            )

            if pd.notna(x)

            and isinstance(
                x,
                (int,float)
            )

            else x

        )

    return df
```

```python
def convert_xlsx_to_json(
    file_path: Path,
    output_dir: Path
):

    print(
        f"Processing {file_path.name}"
    )

    province_code = (
        file_path.stem
        .split("_")[0]
        .strip()
    )

    province_info = (
        PROVINCE_MAPPING
        .get(
            province_code,
            {}
        )
    )

    df = pd.read_excel(
        file_path
    )

    year_cols = []

    for c in df.columns:

        try:

            int(c)

            year_cols.append(c)

        except:

            pass

    result = []

    for _, row in df.iterrows():

        indicator = row.iloc[1]

        if pd.isna(
            indicator
        ):

            continue

        values = {}

        for y in year_cols:

            val = row[y]

            if pd.notna(
                val
            ):

                if isinstance(
                    val,
                    (float, int)
                ):

                    val = round(
                        float(val),
                        4
                    )

                values[
                    str(y)
                ] = val

        item = {

            "id":

            int(row.iloc[0])

            if pd.notna(
                row.iloc[0]
            )

            else None,

            "indicator":

            str(
                indicator
            ).strip(),

            "unit":

            ""

            if pd.isna(
                row.iloc[2]
            )

            else str(
                row.iloc[2]
            ).strip(),

            "values":

            values

        }

        result.append(
            item
        )

    payload = {

        "provinceCode":
            province_code,

        "provinceShortName":
            province_info.get(
                "shortName",
                ""
            ),

        "provinceName":
            province_info.get(
                "name",
                ""
            ),

        "data":
            result

    }

    output_file = (
        output_dir /
        f"{file_path.stem}.json"
    )

    with open(

        output_file,

        "w",

        encoding="utf-8"

    ) as f:

        json.dump(

            payload,

            f,

            ensure_ascii=False,

            indent=2

        )

    print(
        f"Done -> {output_file.name}"
    )
```

```python
clean_xlsx_files = list(
    CLEAN_FOLDER.glob(
        "*.xlsx"
    )
)
```

```python
for file in clean_xlsx_files:

    try:

        convert_xlsx_to_json(

            file,

            OUTPUT_FOLDER

        )

    except Exception as e:

        print(

            file.name,

            e

        )

print(
    "\nCompleted"
)
```
