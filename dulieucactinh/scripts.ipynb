```python
import pandas as pd
import os
import json
from pathlib import Path
```

```python
INPUT_FOLDER = Path(
    r"D:\doanatan\dulieucactinh"
)

OUTPUT_FOLDER = INPUT_FOLDER / "json"

OUTPUT_FOLDER.mkdir(
    exist_ok=True
)

print(
    "Input:",
    INPUT_FOLDER
)

print(
    "Output:",
    OUTPUT_FOLDER
)
```

```python
xlsx_files = list(
    INPUT_FOLDER.glob(
        "*.xlsx"
    )
)

print(
    f"Found {len(xlsx_files)} files"
)

for f in xlsx_files:

    print(
        f.name
    )
```

```python
def convert_xlsx_to_json(
    file_path:Path,
    output_dir:Path
):

    print(
        f"Processing {file_path.name}"
    )

    df = pd.read_excel(
        file_path
    )

    year_cols=[]

    for c in df.columns:

        try:

            int(c)

            year_cols.append(c)

        except:

            pass

    result=[]

    for _,row in df.iterrows():

        indicator=row.iloc[1]

        if pd.isna(
            indicator
        ):

            continue

        values={}

        for y in year_cols:

            val=row[y]

            if pd.notna(
                val
            ):

                if isinstance(
                    val,
                    (float,int)
                ):

                    val=round(
                        float(val),
                        4
                    )

                values[
                    str(y)
                ]=val

        item={

            "id":

            int(row.iloc[0])

            if pd.notna(
                row.iloc[0]
            )

            else None,

            "indicator":

            str(
                indicator
            ).strip(),

            "unit":

            ""

            if pd.isna(
                row.iloc[2]
            )

            else str(
                row.iloc[2]
            ).strip(),

            "values":

            values

        }

        result.append(
            item
        )

    output_file = (
        output_dir /
        f"{file_path.stem}.json"
    )

    with open(

        output_file,

        "w",

        encoding="utf-8"

    ) as f:

        json.dump(

            result,

            f,

            ensure_ascii=False,

            indent=2

        )

    print(
        f"Done -> {output_file.name}"
    )
```

```python
for file in xlsx_files:

    try:

        convert_xlsx_to_json(

            file,

            OUTPUT_FOLDER

        )

    except Exception as e:

        print(

            file.name,

            e

        )

print(
    "\nCompleted"
)
```
